```python
import scipy.stats as sts
from statsmodels.formula.api import ols
import statsmodels.api as statsapi
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

```

ANOVA (Analysis of Variance) is a statistical test used to analyze the difference between the means of more than two groups. The two-way ANOVA compares the mean differences between groups that have been split on two independent variables. Its primary purpose is to understand if there is an interaction between the two independent variables on the dependent variable (statistics.laerd.com, 2020). For this task I have chosen a supermarket sales dataset (https://www.kaggle.com/aungpyaeap/supermarket-sales) which consists of a supermarket company historical sales recorded in 3 different branches for 3 months.

```python
anova=pd.read_csv('./supermarket_sales.csv')

anova.rename(columns={"Product line": "Category","Unit price":"Price"},inplace=True)
anova.dropna(inplace=True)
anova.head(20)
```

```python
def count_outliers(array, m=3):
    return len(array[abs(array - np.mean(array)) > m * np.std(array)])
```

As stated in the definition of two-way ANOVA, we first need to choose two independent variables (also called factors) each with at least two levels and a continuous variable (or the dependent variable) and check whether the dependent variable is affected by the independent variables or by their interaction. For this particular dataset the first factor that I picked is called **Gender** with two levels: **Male and Female** and the second factor is called **Category** and consists of 6 levels: **Health and beauty, Electronic accessories, Home and lifestyle, Sports and travel, Food and beverages, Fashion accessories**.

```python
genders=anova['Gender'].unique()
genders
```

```python
categories=anova['Category'].unique()
categories
```

We want to verify if gender influences the total amount paid by customers, if the category of the products bought has an influence on the total amount paid, and last but not least if there is an interaction between the gender and category affecting the total amount paid. Before moving on, there are a **few assumptions that need to be checked** in order to use two-way ANOVA.
The main and most important assumption needed in order to perform ANOVA is **homogeneity of variance**, this is performed using the **Bartlett test**. Since the Bartlett p value obtained is higher than 5% the null hypothesis that variance is homogenous holds so this assumption is passed.
Another assumption is the **lack of outliers** which is proven by counting the number of outliers, with 6 outliers in all possible groups, so this assumption holds as well.

```python

fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 60))

bartlett_array=[]
right=0
i=0
for prod_line in categories:
    for gender in genders:
        
       

        
        result= anova.loc[(anova['Category']==prod_line) & (anova['Gender']==gender),['Total']].iloc[:,0].values
        axes[i,right].hist(result)
        if(right==1):
            i=i+1
        if (right==0):
            right=1
        else:
            right=0
            
        bartlett_array.append(result)
        
        print("Grouping" + " (" + gender +  " " + prod_line + ")" + " with " + str(len(result)) + " elements has outliers:"+  str(count_outliers(result)))
        
print("\n\nBartlett test for groups: "+str(sts.bartlett(*bartlett_array)) + '\n')
```

Having checked all this assumptions the next step is to apply two-way ANOVA on our data. Two-way ANOVA tests **three null hypotheses**:
<ul>
<li>The means of the continuous variable are equal for different values of the first factor</li>
<li>The means of the continuous variable are equal for different values of the second factor</li>
<li>There is no interaction between the two factors</li>
</ul>
The results after applying two-way ANOVA are illustrated in the below table:

```python
moore_lm = ols('Total ~ C(Gender, Sum)*C(Category, Sum)',
              data=anova).fit()
statsapi.stats.anova_lm(moore_lm, typ=2)
```

Looking at the results we can draw the following conclusions:
<ul>
<li>Gender doesn’t have a significant effect on the price of shopping is accepted although it is close to the 5% threshold (𝐻0)<\li>
<li>Category has no significant effect on the price of shopping is also accepted due to the very large p value, far larger than 5% (𝐻1)</li>
<li>$H_2$ is rejected, the two seem to have an interaction effect, there are categories in which men buy more expensive items than women, and the other way around categories where women buy more expensive items than men</li>
</ul>

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.boxplot(x="Total", y="Category", hue="Gender",
                    data=anova, palette="muted")
```

#### We can see the confirmation of the 𝐻2 in the above boxplot, women tend to spend more on home and lifestyle and food, while men tend to spend more on health and sports.

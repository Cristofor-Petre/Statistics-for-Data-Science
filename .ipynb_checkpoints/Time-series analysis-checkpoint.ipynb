```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARMAResults 
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.arima_model import _arma_predict_out_of_sample
import matplotlib
import numpy as np
import sys
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_squared_error
import pyreadr
import csv
import seaborn as sns
```

#### The dataset is referring to the number of persons in the civilian labour force in Australia each month from Feb 1978 – Aug 1995 and can be found  <a href="https://github.com/robjhyndman/fma">here</a>.

```python
res = pyreadr.read_r(r'labour.rda')
```

```python
res
```

```python
loaded_ds = (dict(res)['labour'])

loaded_ds
```

```python
dates = pd.date_range(start='2/1/1978', end='9/1/1995', freq='M')
loaded_ds['Date'] = dates
```

```python
loaded_ds.head(10)
```

```python
loaded_ds = loaded_ds[loaded_ds.labour.notnull()]

def index_data(dataset):
    dataset = dataset.set_index('Date')
    return dataset

```

```python
timeseries_processed = index_data(loaded_ds)
initial_data_processed_df = timeseries_processed['labour']
```

#### To get a better insight from this dataset, we first plot a line graph:

```python
plt.figure(figsize=(12, 7))
plt.xlabel('date', fontsize=14)
plt.plot(list(loaded_ds.Date.astype(str)), list(initial_data_processed_df),'g')

ax = plt.gca()
temp = ax.xaxis.get_ticklabels() 
temp = list(set(temp) - set(temp[::50]))

for label in temp:
    label.set_visible(False)

plt.legend(loc=4, prop={'size': 14})
plt.title('Number of persons in the civilian labour force in Australia')
```

```python
sns.distplot(list(initial_data_processed_df), kde=False)
```

An important part in the process of fitting a model like ARMA is to check **whether the time series is stationary or not** (since stationarity of the dataset is required to use such models). **A time series is stationary if it doesn’t have a trend or seasonal effects and parameters such as mean and variance are consistent over time**. Our plot clearly shows an ascending trend over time. The seasonal trends are observed from the similar shapes in each month. Next, we focus on methods for making the time series stationary. But before that, we plot the **autocorrelation function** with a maximum number of 45 lags to get better hints of what we could further do.

```python
trash_var = plot_acf(initial_data_processed_df, lags=45)
plt.title('ACF before')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12, 7)
```

We notice a **slow decrease** with each lag meaning we have to use **differencing** to make our dataset stationary. Differencing is performed by subtracting the previous observation from the current observation ($x_{t} - x_{t-1}$) resulting in stabilizing the mean and variance.

```python
copy = timeseries_processed.copy()
diff = True
diff_data = []

if diff:
    diff_data = copy['labour'] - copy['labour'].shift()
else:
    diff_data = copy['labour']
    

diff_data[0] = 0
```

```python
diff_data
```

#### After performing differencing, we plot a line graph of the obtained differentiated values and we can clearly see that the values **are centered around the mean**, making the dataset more stationary.

```python
plt.figure(figsize=(12, 7))
plt.xlabel('', fontsize=14)
plt.ylabel('', fontsize=14)
plt.plot(diff_data,'g')

ax = plt.gca()
temp = ax.xaxis.get_ticklabels() 
temp = list(set(temp) - set(temp[::200]))

for label in temp:
    label.set_visible(False)

plt.legend(loc=4, prop={'size': 14})
```

#### After performing differencing, having obtained a more stationary dataset, the next step is to look at autocorrelation (ACF) and partial autocorrelation (PACF) plots since they are crucial in deciding what models to choose.

```python
trash_var = plot_pacf(diff_data, lags=25)
plt.title('PACF after differentiation')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12, 7)
```

```python
trash_var = plot_acf(diff_data, lags=25)
plt.title('ACF after differentiation')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12, 7)
```

#### Visually inspecting the above graphs, some possible candidates for the MA parameter (given by the ACF graph) could be 1, 12 or 24 while for the AR parameter (given by PACF) we could consider the values 1, 2 or 7.

```python
first_model=ARMA(diff_data, order = (1,1)).fit()
first_model_predict = first_model.predict(0)

print ('ARMA('+str(1)+','+str(1)+')'+' with AIC:'+str(first_model.aic)+' and R2:'+str(r2_score(diff_data,first_model_predict)))

second_model=ARMA(diff_data, order = (1,12)).fit()
second_model_predict = second_model.predict(0)

print ('ARMA('+str(1)+','+str(12)+')'+' with AIC:'+str(second_model.aic)+' and R2:'+str(r2_score(diff_data,second_model_predict)))

third_model=ARMA(diff_data, order = (1,4)).fit()
third_model_predict = third_model.predict(0)

print ('ARMA('+str(1)+','+str(4)+')'+' with AIC:'+str(third_model.aic)+' and R2:'+str(r2_score(diff_data,third_model_predict)))



fourth_model=ARMA(diff_data, order = (2,11)).fit()
fourth_model_predict = fourth_model.predict(0)

print ('ARMA('+str(2)+','+str(11)+')'+' with AIC:'+str(fourth_model.aic)+' and R2:'+str(r2_score(diff_data,fourth_model_predict)))



fifth_model=ARMA(diff_data, order = (2,3)).fit()
fifth_model_predict = fifth_model.predict(0)

print ('ARMA('+str(2)+','+str(3)+')'+' with AIC:'+str(fifth_model.aic)+' and R2:'+str(r2_score(diff_data,fifth_model_predict)))

```

```python
model = fourth_model
```

```python
result_predicted = model.predict(0)
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12, 7)
plt.plot(result_predicted, 'b')
plt.plot(diff_data, 'g')
plt.title("predicted and initial values")
plt.xticks(rotation=90)
ax = plt.gca()
temp = ax.xaxis.get_ticklabels() 
temp = list(set(temp) - set(temp[::50]))
for label in temp:
    label.set_visible(False)
```

```python
pred_reverse = [copy['labour'][0]]
i = 1
while(i < len(result_predicted)):
    pred_reverse.append(pred_reverse[i-1] + result_predicted[i])
    i= i + 1
```

#### After choosing a model, the next step is to compare the original time series against the predictions made by our model. Looking at the below figure, the predictions made by the model are not perfect, however we can see the tendency preserved.

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12, 7)
plt.plot(pred_reverse, 'y', label='predicted values')
plt.plot(list(copy['labour']), 'g', label='true values')

ax = plt.gca()
temp = ax.xaxis.get_ticklabels()
temp = list(set(temp) - set(temp[::50]))
plt.legend(loc="upper left")
#for label in temp:
#    label.set_visible(False)
    
model.aic
```

```python
days_to_predict = 20

next_20_days = _arma_predict_out_of_sample(model.params, days_to_predict, model.resid, model.k_ar, model.k_ma, model.k_trend, model.k_exog, endog=diff_data, exog=None, start=len(diff_data))
```

```python
next_20_days
```

```python
predicted_ = []
predicted_.append(timeseries_processed['labour'][x] + next_20_days[0])
```

```python
for i in range(1, days_to_predict):
    predicted_.append(predicted_[i+x] + next_20_days[i])
```

```python
forecast_days = np.arange(len(timeseries_processed), len(timeseries_processed) + len(predicted_) +1)
```

```python
predicted_.insert(0, timeseries_processed['labour'][x])
```

#### And finally our next task is to forecast the next 20 steps by predicting on the differentiated values using the function _arma_predict_out_of_sample from the statsmodels library. Looking at the below figure we can again observe that our predictions preserved some of the general tendency of the original data.

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12, 7)

plt.plot(timeseries_processed['labour'].values, label = 'values')
plt.plot(forecast_days, predicted_, label = 'prediction')
plt.legend(loc="upper left")
```
